```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
goruntu=cv2.imread("with color.jpg",0)
equ=cv2.equalizeHist(goruntu)
# cv2_imshow expects only the image as an argument
cv2_imshow(goruntu)
cv2_imshow(equ)
```

```python
axl=plt.subplot(211)
# Assuming ax1 was a typo and you meant axl
axl.set_title("orijinal görüntü histogramı",fontsize=20)
plt.hist(goruntu.flatten(),256,[0,256], color = 'r')
ax2=plt.subplot(212)
ax2.set_title("eșitlenmiş görüntü histogramı",fontsize=20)
ax2.hist(equ.flatten(),256,[0,256], color ='b')
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
equ=clahe.apply (goruntu)
cv2_imshow (goruntu)
cv2_imshow(equ)
axi=plt.subplot (221)
axl.set_title ("orijinal görüntü histogramı", fontsize=20)
plt.hist (goruntu.flatten(),256,[0,250], color = 'r')
ax2=plt.subplot(222)
ax2.set_title("eșitlenmiş görüntü histogramın", fontsize=20)
ax2.hist(equ.flatten(),256,[0,256], color='b')
plt.show()
```
