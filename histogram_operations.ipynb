```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
renkli_goruntu = plt.imread("Gemini_Generated_Image_gr30xzgr30xzgr30.jpg")
color = ['blue','green','gray','red']
```

```python
gri_tonlamali = cv2.imread("Gemini_Generated_Image_gr30xzgr30xzgr30.jpg",0)
```

```python
for i,col in enumerate(color):
    print(col)
    if i<3:
      histogram = cv2.calcHist([renkli_goruntu],[i],None,[256],[0,256])
    if i==3:
      histogram = cv2.calcHist([gri_tonlamali],[0],None,[256],[0,256])
    plt.plot(histogram,color = col)
    plt.xlim([0,256])
plt.show()
```
